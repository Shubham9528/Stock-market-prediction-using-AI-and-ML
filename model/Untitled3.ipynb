```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
ss=pd.read_csv("NIFTY.csv")
```

```python
ss
```

```python
ss['Date'] = pd.to_datetime(ss['Date'])
```

```python
ss['Year'] = ss['Date'].dt.year
ss['Month'] =ss['Date'].dt.month
ss['Day'] = ss['Date'].dt.day
ss['dayOfWeek'] = ss['Date'].dt.day_name()
ss['WeekOfYear'] = ss['Date'].dt.weekofyear

ss.head(419)
```

```python
y=ss['Close']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ss['Indices']=le.fit_transform(ss['Indices'].astype("str"))
ss['Indices'].value_counts()
```

```python
X = ss[['Day','Month','Year','Indices']]
```

```python
poly2 = PolynomialFeatures(degree =2)
X_poly = poly2.fit_transform(X)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=15, random_state=0)
regressor.fit(X_poly,y)
```

```python
print(regressor.score(X_poly, y))
```

```python
Day = 26


Month =2
Year =2021
Indices=2

#Samples = Samples_31_July[0,0]
X1 = np.array([[Day,Month,Year,Indices]])
Total_infections  = regressor.predict(poly2.fit_transform(X1))
```

```python
print('Total No of cases -> {}'.format(int(Total_infections[0])))
```
